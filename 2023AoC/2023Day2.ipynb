```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("input.txt", sep='\r', lineterminator='\n', header=None)
df
```

```python
ndf = df[0].str.split(';',expand=True)
#ndf[1] = ndf[1].str.split(';',expand=True)
ndf[['Game',0]] = ndf[0].str.split(':',expand=True)
ndf
```

```python
blueIndex = pd.DataFrame()
for column in ndf:
    blueIndex[column] = ndf[column].str.find("blue")


greenIndex = pd.DataFrame()
for column in ndf:
    greenIndex[column] = ndf[column].str.find("green")


redIndex = pd.DataFrame()
for column in ndf:
    redIndex[column] = ndf[column].str.find("red")
```

```python
blueValues = pd.DataFrame()
    

for col in ndf.columns:
    a = []
    for row in range(0, len(ndf)):
        b = ndf[col][row]
        c = blueIndex[col][row] -3
        if np.isnan(c) or c < 0:
             a.append(0)
        else: 
            d = b[int(c):int(c)+2]
            a.append(d) 
    blueValues[col] = a


redValues = pd.DataFrame()
    

for col in ndf.columns:
    a = []
    for row in range(0, len(ndf)):
        b = ndf[col][row]
        c = redIndex[col][row] -3
        if np.isnan(c) or c < 0:
             a.append(0)
        else: 
            d = b[int(c):int(c)+2]
            a.append(d) 
    redValues[col] = a


greenValues = pd.DataFrame()
    

for col in ndf.columns:
    a = []
    for row in range(0, len(ndf)):
        b = ndf[col][row]
        c = greenIndex[col][row] -3
        if np.isnan(c) or c < 0:
             a.append(0)
        else: 
            d = b[int(c):int(c)+2]
            a.append(d) 
    greenValues[col] = a
blueValues
```

```python
#Part One
blue = 14
red = 12
green = 13

listToSum = []

#print(blueValues)

for index, row in blueValues.iterrows():
    a = 0
    for col in blueValues.columns:
        element = int(row[col])
        if element <= blue:
            a = a + 0
        else:
            a = a + 1
    if a > 0:
        b = index +1
        listToSum.append(b)
    else: 
        listToSum.append(0) 

for index, row in redValues.iterrows():
    a = 0
    for col in redValues.columns:
        element = int(row[col])
        if element <= red:
            a = a+0
        else:
            a = a +1
    if a > 0:
        b = index +1
        listToSum.append(b)
    else: 
        listToSum.append(0)  

for index, row in greenValues.iterrows():
    a = 0
    for col in greenValues.columns:
        element = int(row[col])
        if element <= green:
            a = a+0
        else:
            a = a +1
    if a > 0:
        b = index +1
        listToSum.append(b)
    else: 
        listToSum.append(0) 

print(listToSum)
test_list = list(set(listToSum))
print(test_list)
            

```

```python
#part 2

maxBlues = []

for index, row in blueValues.iterrows():
    a = []
    for col in blueValues.columns:
        element = int(row[col])
        a.append(element)
    maxBlues.append(max(a))

maxReds = []

for index, row in redValues.iterrows():
    a = []
    for col in redValues.columns:
        element = int(row[col])
        a.append(element)
    maxReds.append(max(a))


maxGreens = []

for index, row in greenValues.iterrows():
    a = []
    for col in greenValues.columns:
        element = int(row[col])
        a.append(element)
    maxGreens.append(max(a))
        

maxes = pd.DataFrame(list(zip(maxBlues,maxReds,maxGreens)))

maxes['Powers'] = maxes[0] *maxes[1]*maxes[2]
print(sum(maxes['Powers']))
```
